```python
# imgaug invoice augmentation
import cv2
import imgaug.augmenters as iaa
import glob
import os

datasets = []

img_path = glob.glob("D:/images/yolov5/invoices/normalized/*.jpg")
print(img_path)

for images in img_path:
    img = cv2.imread(images)
    datasets.append(img)

augmentation = iaa.Sequential([
    # iaa.Rotate((-30, 30))  
    iaa.Fliplr(0.5),
    iaa.Flipud(0.5),
    
    iaa.Affine(
        translate_percent={"x":(-0.2, 0.2), "y":(-0.2, 0.2)},
        rotate=(-30, 30),
        scale ={"x":(0.5, 1.5), "y":(0.5, 1.5)}
    ),
    
    iaa.Multiply((0.8, 1.2)),
    
    iaa.LinearContrast((0.6, 1.4)),
    
    iaa.GaussianBlur((0.0, 3.0))
])

augmented_img = augmentation(images=datasets)

for invoices in augmented_img:
    # print(invoices)
    cv2.imshow("Invoices", invoices)
    cv2.imwrite("affine.jpg", invoices)
    cv2.waitKey(0)
    
    
```

```python
# invoice rotation augmentation using imgaug

import glob
import cv2
import numpy as np
import imgaug.augmenters as iaa
import os

path = glob.iglob('D:/images/yolov5/invoices/normalized/**/*', recursive=True)

datasets = []

for files in path:
    print(files)
    img = os.path.basename(files)
    print(img)
    # image = cv2.imread("D:/images/yolov5/invoices/BESCOM Jan2021.jpg")
    image = cv2.imread(files)
    datasets.append(image)
    
    augmentation = iaa.Sequential([ 
        # iaa.Fliplr(0.5),
        # iaa.Flipud(0.5)
        iaa.Affine(rotate=(-20, 20), fit_output=True)
    ])

    augmented_img = augmentation(images=datasets)

    for invoices in augmented_img:
        # print(invoices)
        # cv2.imshow("Invoices", invoices)
        cv2.imwrite("rotation_" + img, invoices)
        cv2.waitKey(0)

   
```

```python
# affine translation with 40 and 60 pixel on x and y axis respectively
import glob
import os
import cv2
import numpy as np
import imgaug.augmenters as iaa

path = glob.iglob('D:/images/normalizedinvoices/totalinvoices/**/*.jpg', recursive=True)

datasets = []

for files in path:
    img = os.path.basename(files)
    print(img)
    
    image = cv2.imread(files)
    datasets.append(image)
    
    augmentation = iaa.Sequential([
       iaa.Affine(
        translate_px={"x": 40, "y": 60},
        scale=(0.5, 0.7))
    ])
    
    augmented_img = augmentation(images=datasets)
    
    for invoices in augmented_img:
        cv2.imwrite("transformation_" + img, invoices)
        
    
```
